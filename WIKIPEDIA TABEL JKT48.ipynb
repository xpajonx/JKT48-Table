```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = requests.get("https://id.wikipedia.org/wiki/Daftar_anggota_JKT48")
```

```python
soup = BeautifulSoup(url.content,'lxml')
```

```python
#scraping tabel team j
table_team_j = soup.find_all('table',class_="wikitable sortable")[1:5]
table_team_j1 = pd.read_html(str(table_team_j),attrs=list)
table_team_j2 = table_team_j1[0].to_dict()
table_team_j3 = pd.DataFrame(table_team_j2)
```
