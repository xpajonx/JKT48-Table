```python
# merapikan kolom tempat lahir
row = 0
for i in jkt48_frame['Kelahiran_1']:
    word = []
    for char in i:
        if char == ',':
            break
        else:
            word.append(char)
    s = ''.join(word)
    jkt48_frame['Kelahiran_1'][row] = s
    row += 1
```
