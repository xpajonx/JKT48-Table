```python
def table_generator(site):
    url = requests.get(site)
    soup = BeautifulSoup(url.content,'lxml')
    clean_soup = soup.find_all('table',class_='wikitable')
    clean_list = pd.read_html(str(clean_soup),attrs=list)
    fix_table = pd.concat(clean_list)
    return fix_table
```

```python
# iterasi semua tahun ke dalam masing2 variabel
c=0
for i in list_website:
    locals()['group'+str(c)] = table_generator(i)
    c += 1
    print ('test')
```
